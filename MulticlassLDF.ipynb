# Classification with Label Dependent Features

In the [MulticlassClassification]() notebook, we saw how to build classifiers whose output is one of K classes. In those examples, each class was treated as an atomic unit: class 1 had no more to do with class 2 than it did with class 65. In some cases, you want to associate features with the different classes (or labels): this gives rise to **label dependent feature** models. (These are also useful to [zero-shot learning](#zeroshot), which we'll see later in this notebook.)

## Formalism and Data Format

A natural problem where LDF arises is in search ranking-like problems. Here, the input to the system is a query (like "vowpal wabbit ldf howto") and the output is one of the billion-or-so documents on the web. It doesn't make a lot of sense to give each document on the web a unique id and then try to learn a mapping from inputs (queries) to outputs (documents), treating all documents independently. Instead, what you *want* to do is associate a feature vector with a *pair* $(x,y)$, where $x$ is the query and $y$ is a hypothetical returned document. This feature vector might do things like count how many of the query terms are present in the corresponding document.

The associated LDF learning problem is: given some $x$ (query) and a set of possible retreived documents $(y_1, y_2, \dots, y_K)$, we want to compute $\arg\min_k f( \varphi(x,y_k) )$, where $\varphi(x,y_k)$ is the features you extract for that pair, and $f$ is the learned "cost" function. *[Note: sometimes people talk about a "score" function that we try to maximize rather than a "cost" function that we try to minimize. Here, we'll talk only about costs, so lower is always better.]*

The basic data format (we'll see briefer alternatives later) for LDF is something like the following:

    1:1 | features for pair (x1,y1) ...
    2:0 | features for pair (x1,y2) ...  <-- this is the correct "label"
    3:1 | features for pair (x1,y3) ...
    4:1 | features for pair (x1,y4) ...
    <blank line>
    1:0 | features for pair (x2,y1) ...  <-- this is the correct "label"
    2:1 | features for pair (x2,y2) ...
    5:1 | features for pair (x2,y5) ...
    <blank line>
    
And so on. Here, each **block** of data is separated by a blank line. A single block corresponds to a single training example, and each line in that block corresponds to a valid prediction for that example. The "label" portion consists of a class identifier (1, 2, 3, 4, 5, ...) and a cost (:1 means "cost=1" and :0 means "cost=0"). After that, any features that make sense can be written out.

A few notes:

1. Costs can be any non-negative values, and there can be multiple labels with a cost of zero.
1. The actual labels (1, 2, 3, 4, 5, ...) are completely ignored by the learning algorithm. Everything you want to encode *must* be encoded in the features. If each line in a single block has exactly the same features, it will be impossible to differentiate them. This also means that label ids needn't be consistent across blocks.
1. You don't have to use consecutive label ids (as in the second block above).

## LDF for Quizbowl, Step 1

I'm going to assume you've already gone through the [MulticlassClassification]() notebook and have downloaded the QuizBowl data. If not, do so now. Here is some code copied from before for loading the data.

```python
import csv, sys, re, random, os

def sanify(s): return s.replace(':',';').replace('|','/')   # replace : with ; and | with /
def tokenize(s): return re.sub('([^A-Za-z0-9 ]+)', ' \\1 ', s).split()  # add space around anything not alphanum

def readQuizBowlData(filename):
    train,dev,test = [],[],[]
    data = csv.reader( open(filename, 'r').readlines() )
    header = data.next()
    if header != ['Question ID', 'Fold', 'Category', 'Answer', 'Text']:
        raise Exception('data improperly formatted')
    for item in iter(data):
        y = item[3]        
        x = sanify(' '.join(tokenize(item[4].replace('|||',''))))
        if   item[1] == 'train': train.append( (x,y) )
        elif item[1] == 'dev'  :   dev.append( (x,y) )
        elif item[1] == 'test' :  test.append( (x,y) )
    return train,dev,test

def makeLabelIDs(train, outputFile):
    labelIds = { label: k+1 for k,label in enumerate(set([y for x,y in train])) }
    labelIds['***UNKNOWN***'] = len(labelIds)+1
    with open(outputFile, 'w') as h:
        for label,k in labelIds.iteritems():
            print >>h, '%d\t%s' % (k,label)
    return labelIds

train,dev,test = readQuizBowlData('data/questions/questions.csv')
```


We're going to start off by basically replicating the experiment there, but using a very wasteful LDF format. In particular, for each quizbowl example, we'll generate a single block. This block will have one line *for every class*. For class 10, for instance, we'll replace every features "f" with "f_10". This will be horribly bloated, but we'll fix that later.

In order to actually make this not take huge amounts of time and disk space, we'll cut down the train/dev/test significantly (to the point that they're toy data).

```python
random.seed(9876)
random.shuffle(train)
train_small = train[:10]
labelIds_small = makeLabelIDs(train_small, 'data/quizbowl.ldf.labels')
print 'total of %d labels, including one for unknown' % len(labelIds_small)

def writeVWFile(filename, data, labelIds):
    unknownId = labelIds['***UNKNOWN***']
    with open(filename,'w') as h:
        for x,y in data:
            trueLabel = labelIds.get(y, unknownId)
            feats     = x.split()
            # generate a single block by going over EVERY POSSIBLE LABEL
            for k in range(1, unknownId+1):           # unknownId is the last label
                cost = 0. if trueLabel == k else 1.   # how bad is it to predict k?
                x_k  = ' '.join(['%s_%d' % (f,k) for f in feats])  # example for label k
                print >>h, '%d:%g |q %s' % (k, cost, x_k)
            print >>h, ''   # blank line to separate blocks
            
writeVWFile('data/quizbowl.ldf.tr', train_small, labelIds_small)
!wc data/quizbowl.ldf.tr
!head -n12 data/quizbowl.ldf.tr
```

So in this silly data, we have 11 labels, so all blocks has length 11. One of these (in this case, label 8) is the correct label and has a cost of zero.

Before we worry about shrinking this file, let's train and test a model on it. The main thing is the `--csoaa_ldf multiline` argument, which says we want to run cost-sensitive one against all training, where data is provided in multiline format.

```python
!vw -k -c --csoaa_ldf multiline -d data/quizbowl.ldf.tr --passes 10 --holdout_off -f data/quizbowl.ldf.model
!vw -i data/quizbowl.ldf.model -t -d data/quizbowl.ldf.tr -r data/quizbowl.ldf.tr.raw
!head -n24 data/quizbowl.ldf.tr.raw
```

Here, we've trained a model and made predictions on the training data. The numbers printed in the raw file are the expected costs.

Now, there's a lot of stuff going on here that's silly. We'll address these in order:

1. The features in the above example are really a cross product between the input $x$ and the label $y_k$. But `vw` has built in support for cross-produces with `-q`, so let's just use that. After fixing that problem, the same input $x$ gets repeated over and over again. That's really redundant.
1. Now, the same features (or in this case, single feature) gets repeated over and over again for the label. That's also redundant.
1. Finally, we'd like more interesting features on the label.

## Separating out the input and output using quadratic features: shared features

Hopefully it's clear that a training example like:

    1:1 |q If_1 it_1 were_1 always_1 true_1 ,_1 fractional_1 distillation_1 would_1 ...
    
is exactly the same as:

    1:1 |l 1 |q If it were always true , fractional distillation would ...

if `-q lq` is added to the command line. This simplifies the data, but then the blocks will look like:

    1:1 |l 1 |q If it were always true , fractional distillation would ...
    2:1 |l 2 |q If it were always true , fractional distillation would ...
    3:1 |l 3 |q If it were always true , fractional distillation would ...
    4:1 |l 4 |q If it were always true , fractional distillation would ...
    ...
    
where the "`q`" part will be the same in every example. We can fix this problem by having "`shared`" features for blocks. This looks like the following:

    ***shared*** |q If it were always true , fractional distillation would ...
    1:1 |l 1
    2:1 |l 2
    3:1 |l 3
    4:1 |l 4
    ...

What this says is: pretend that the `shared` line appears on every example within this block. We can fix our code to output data that looks like this, which will be much smaller.

```python
def writeVWFile(filename, data, labelIds):
    unknownId = labelIds['***UNKNOWN***']
    with open(filename,'w') as h:
        for x,y in data:
            print >>h, 'shared |q %s' % x
            trueLabel = labelIds.get(y, unknownId)
            # generate a single block by going over EVERY POSSIBLE LABEL
            for k in range(1, unknownId+1):           # unknownId is the last label
                cost = 0. if trueLabel == k else 1.   # how bad is it to predict k?
                print >>h, '%d:%g |l %d' % (k, cost, k)
            print >>h, ''   # blank line to separate blocks
            
writeVWFile('data/quizbowl.ldf.tr', train_small, labelIds_small)
!head -n13 data/quizbowl.ldf.tr
!vw -k -c --csoaa_ldf multiline -d data/quizbowl.ldf.tr --passes 10 --holdout_off -f data/quizbowl.ldf.model -q lq
!vw -i data/quizbowl.ldf.model -t -d data/quizbowl.ldf.tr -r data/quizbowl.ldf.tr.raw
!head -n24 data/quizbowl.ldf.tr.raw
```

Note that the output is similar but not identical. It's not identical because there are extra features (the `-q` continues to include the linear terms, which are not in the original formulation).

## Headers with label-specific features

In some cases, you really want to be able to define whatever arbitrary features you want for $\varphi(x,y_k)$. But in many cases, $\varphi(x,y_k)$ is just the outer product of some features of $x$ (that might be written as `shared`) features and some features of $y_k$. In the case above, the features of $y_k$ is just the singleton feature `k`, but (as we'll see below) you often want more. We can specify all of these in a header as follows:

    label:1 |l 1
    label:2 |l 2
    ...
    label:11 |l 11
    <blank line>
    
This batch of label definitions basically says "any time you see an example with, eg, label 2, add "`|l 2`" to it. We can then generate the equivalent data much more simply as:

```python
def writeVWFile(filename, data, labelIds):
    unknownId = labelIds['***UNKNOWN***']
    with open(filename,'w') as h:
        for k in range(1, unknownId+1):
            print >>h, 'label:%d |l %d' % (k,k)
        print >>h, ''
        
        for x,y in data:
            print >>h, 'shared |q %s' % x
            trueLabel = labelIds.get(y, unknownId)
            # generate a single block by going over EVERY POSSIBLE LABEL
            for k in range(1, unknownId+1):           # unknownId is the last label
                cost = 0. if trueLabel == k else 1.   # how bad is it to predict k?
                print >>h, '%d:%g | bias' % (k, cost)
            print >>h, ''   # blank line to separate blocks
            
writeVWFile('data/quizbowl.ldf.tr', train_small, labelIds_small)
!echo -n "\nTRAINING DATA:\n\n"
!head -n20 data/quizbowl.ldf.tr
!echo -n "\nTRAINING:\n\n"
!vw -k -c --csoaa_ldf multiline -d data/quizbowl.ldf.tr --passes 10 --holdout_off -f data/quizbowl.ldf.model -q lq
!echo -n "\nTESTING:\n\n"
!vw -i data/quizbowl.ldf.model -t -d data/quizbowl.ldf.tr -r data/quizbowl.ldf.tr.raw
!echo -n "\nOUTPUT (HEAD):\n\n"
!head -n24 data/quizbowl.ldf.tr.raw
```

Note that in the above we had to add a "bias" feature to each example. Technically this should be unnecessary, but for some reason there's an arguable bug in vw where it ignores these examples with no features; until that bug is fixed, you have to have at least *something* written as a feature for each example.

## Actually putting useful features in the label-dependent features

Right now we're not actually sharing any useful information in the label dependent features, *but we can*! For the QuizBowl data, each label corresponds to a Wikipedia page, so we'll make the labels depend on the words that appear in that page.

```python
from collections import Counter

def readWikipedia(labelIds):
    wiki = {}
    directory = 'data' + os.sep + 'questions' + os.sep + 'wiki' + os.sep
    for labelName in labelIds.iterkeys():
        try:
            with open(directory + labelName + '.txt', 'r') as h:
                bow = Counter()
                for l in h.readlines():
                    if not l.startswith('== '): continue
                    for w in tokenize(l.lower())[1:-1]:
                        bow[w] += 1
                wiki[labelName] = ' '.join(['%s:%d' % (sanify(w), c) for (w,c) in bow.iteritems()])
        except IOError:
            print 'warning: could not read wikipedia page for %s' % labelName
            wiki[labelName] = 'unknown'
    return wiki

def writeVWFile(filename, data, labelIds, wikipedia):
    unknownId = labelIds['***UNKNOWN***']
    with open(filename,'w') as h:
        for labelName, k in labelIds.iteritems():
            print >>h, 'label:%d |l label%d %s' % (k, k, wikipedia[labelName])
        print >>h, ''
        
        for x,y in data:
            print >>h, 'shared |q %s' % x
            trueLabel = labelIds.get(y, unknownId)
            # generate a single block by going over EVERY POSSIBLE LABEL
            for k in range(1, unknownId+1):           # unknownId is the last label
                cost = 0. if trueLabel == k else 1.   # how bad is it to predict k?
                print >>h, '%d:%g | bias' % (k, cost)
            print >>h, ''   # blank line to separate blocks
            
labelIds = makeLabelIDs(train, 'data/quizbowl.ldf.labels')
print 'total of %d labels, including one for unknown' % len(labelIds)

wikipedia = readWikipedia(labelIds_small)

writeVWFile('data/quizbowl.ldf.tr', train_small, labelIds_small, wikipedia)
!echo -n "\nTRAINING DATA:\n\n"
!head -n20 data/quizbowl.ldf.tr
!echo -n "\nTRAINING:\n\n"
!vw -k -c --csoaa_ldf multiline -d data/quizbowl.ldf.tr --passes 10 --holdout_off -f data/quizbowl.ldf.model -q lq
!echo -n "\nTESTING:\n\n"
!vw -i data/quizbowl.ldf.model -t -d data/quizbowl.ldf.tr -r data/quizbowl.ldf.tr.raw
!echo -n "\nOUTPUT (HEAD):\n\n"
!head -n24 data/quizbowl.ldf.tr.raw
```
